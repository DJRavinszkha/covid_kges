```python
#generate RDF links for nodes

#1 http://bio2rdf.org/drugbank:
#2 http://bio2rdf.org/drugbank_resource:
#3 http://bio2rdf.org/chebi:
#4 http://bio2rdf.org/mesh: 
#5 http://bio2rdf.org/atc
#6 https://www.ncbi.nlm.nih.gov/gene/

# Create a file that contains all URI's of entities with their identifier

# drugbank = all rows with #1
# drugbank_resource = all rows with 2 
# chebi = all rows with CHEBI
# mesh = all rows with MESH
# gene = all rows with G:


drugbank = DBGN_nodes[DBGN_nodes[0].str.contains("http://bio2rdf.org/drugbank:")]
drugbank_resource = DBGN_nodes[DBGN_nodes[0].str.contains("http://bio2rdf.org/drugbank_resource:")]
chebi = DBGN_nodes[DBGN_nodes[0].str.contains("CHEBI:")]
mesh = DBGN_nodes[DBGN_nodes[0].str.contains("MESH:")]
gene = DBGN_nodes[DBGN_nodes[0].str.contains("G:")]
atc = DBGN_nodes[DBGN_nodes[0].str.contains("http://bio2rdf.org/atc:")]
```

```python
chebi = pd.Series(chebi[0])
chebi = chebi.str.split(":", expand = True)
chebi[3] = "<http://bio2rdf.org/chebi:"
chebi.columns = ("rdfs:type", "chebi", "ID", "URI")
chebi['URI'] = chebi['URI'].astype(str) + chebi['ID'] + ">"
chebi['ID'] = "chebi_" + chebi['ID'].astype(str)
chebi = chebi[['URI', 'ID']]
chebi['datasource'] = 'ChEBI'

mesh = pd.Series(mesh[0])
mesh = mesh.str.split(":", expand = True)
mesh.columns = ("rdfs:type", "mesh", "ID", "URI")
mesh['URI'] = "<http://bio2rdf.org/mesh:"
mesh['URI'] = mesh['URI'].astype(str) + mesh['ID'] + ">"
mesh['ID'] = "mesh_" + mesh['ID'].astype(str)
mesh = mesh[['URI', 'ID']]
mesh['datasource'] = 'MeSH'


gene = pd.Series(gene[0])
gene = gene.str.replace(r"\(.*\)","")
gene = gene.str.split(":", expand = True)
gene[2] = "<https://www.ncbi.nlm.nih.gov/gene/"
gene.columns = ("rdfs:type", "ID", "URI")
gene['URI'] = gene['URI'].astype(str) + gene['ID'] + ">"
gene['ID'] = "NCBI_" + gene['ID'].astype(str)
gene = gene[['URI', 'ID']]
gene['datasource'] = 'Entrez'


drugbank = pd.Series(drugbank[0])
drugbank = drugbank.str.replace(r">","")
drugbank = drugbank.str.split("drugbank:", expand = True)
drugbank[0] = "<http://bio2rdf.org/drugbank:"
drugbank.columns = ("URI", "ID")
drugbank['URI'] = drugbank['URI'].astype(str) + drugbank['ID'] + ">"
drugbank['ID'] = "DB" + drugbank['ID'].astype(str)
drugbank['datasource'] = 'DrugBank'


drugbank_resource = pd.Series(drugbank_resource[0])
drugbank_resource = drugbank_resource.str.replace(r">","")
drugbank_resource = drugbank_resource.str.split("drugbank_resource:", expand = True)
drugbank_resource[0] = "<http://bio2rdf.org/drugbank_resource:"
drugbank_resource.columns = ("URI", "ID")
drugbank_resource['URI'] = drugbank_resource['URI'].astype(str) + drugbank_resource['ID'] + ">"
drugbank_resource['ID'] = "drugbank_resource_" + drugbank_resource['ID'].astype(str)
drugbank_resource['datasource'] = 'DrugBank Resource'


atc = pd.Series(atc[0])
atc = atc.str.replace(r">","")
atc = atc.str.split("atc:", expand = True)
atc[0] = "<https://bioportal.bioontology.org/ontologies/ATC/"
atc.columns = ("URI", "ID")
atc['URI'] = atc['URI'].astype(str) + atc['ID'] + ">"
atc['ID'] = "atc_" + atc['ID'].astype(str)
atc['datasource'] = 'Anatomical Therapeutic Chemical code'

```

```python
DBGN_Nodes_new = pd.concat([chebi, mesh, gene, atc, drugbank, drugbank_resource])
```

#This contains knowledge such as shared identifiers. Both graphs contain the same entities with different identifiers. Therefore for machine learning we must map these identifiers to unify them (see mappingInChiKeys.ipynb). Nevertheless, in the RDF graph, as many identifiers will be mapped to each other as possible. For example, diseases will have rdf:type disease etc. 

```python
#What we have to do here is highlight the types and potential labels for ...
    #... the entities to make querying the RDF graph easier.

#Because DBGN labelled diseases with a D:, compounds with a C: and Genes with a G:, ...
    #... we can easily create triples where the rdf:type is the predicate.
#Remember this is only for the RDF, not for the machine learning
DBGN_DCG = DBGN_nodes.copy()
#Create relations such that anything:  
    # Head D: 
        # Relation <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> 
            # Tail <http://purl.obolibrary.org/obo/DOID_4> 

D = DBGN_DCG[DBGN_DCG[0].str.contains("D:")]
D.columns = ("h", "r")
D["r"] = "a"
D["t"] = "Disease"
D["r_uri"] = "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>"
D["t_uri"] = "<http://purl.obolibrary.org/obo/DOID_4>"
D['h'] = D['h'].str.replace("MESH:", "mesh_")
D['h'] = D['h'].str.replace("CHEBI:", "chebi_")
D['h'] = D['h'].str.replace("G:", "NCBI_")
D['h'] = D['h'].str.replace("C:", "")
D['h'] = D['h'].str.replace("D:", "")
D = D.merge(DBGN_Nodes_new, left_on='h', right_on='ID')
D = D.drop(['ID'], axis = 1)
D.columns = ("h", "r", "t", "r_URI", "t_URI", "h_URI", "h_datasource")
            
    # Head C: 
        # Relation <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> 
            # Tail <https://bio2rdf.org/chebi:23888>    
C = DBGN_DCG[DBGN_DCG[0].str.contains("C:")]
C.columns = ("h", "r")
C["r"] = "a"
C["t"] = "Compound"
C["r_uri"] = "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>"
C["t_uri"] = "<https://bio2rdf.org/chebi:23888>"
C['h'] = C['h'].str.replace("MESH:", "mesh_")
C['h'] = C['h'].str.replace("CHEBI:", "chebi_")
C['h'] = C['h'].str.replace("G:", "NCBI_")
C['h'] = C['h'].str.replace("C:", "")
C['h'] = C['h'].str.replace("D:", "")
C = C.merge(DBGN_Nodes_new, left_on='h', right_on='ID')
C = C.drop(['ID'], axis = 1)
C.columns = ("h", "r", "t", "r_URI", "t_URI", "h_URI", "h_datasource")

    # Head G: 
        # Relation <http://www.w3.org/1999/02/22-rdf-syntax-ns#type> 
            # Tail <https://semanticscience.org/resource/SIO_010035>                
G = DBGN_DCG[DBGN_DCG[0].str.contains("G:")]
G.columns = ("h", "r")
G["r"] = "a"
G["t"] = "Gene"
G["r_uri"] = "<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>"
G["t_uri"] = "<https://semanticscience.org/resource/SIO_010035>"
G['h'] = G['h'].str.replace("MESH:", "mesh_")
G['h'] = G['h'].str.replace("CHEBI:", "chebi_")
G['h'] = G['h'].str.replace("G:", "NCBI_")
G['h'] = G['h'].str.replace("C:", "")
G['h'] = G['h'].str.replace("D:", "")
G = G.merge(DBGN_Nodes_new, left_on='h', right_on='ID')
G = G.drop(['ID'], axis = 1)
G.columns = ("h", "r", "t", "r_URI", "t_URI", "h_URI", "h_datasource")
```
